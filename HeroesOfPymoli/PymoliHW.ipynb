```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
Totalplayer = len(purchase_data_df["SN"].unique())
PlayerCount_df = pd.DataFrame({"Total Players": [Totalplayer]})
PlayerCount_df
```

```python
NumUnique = len(purchase_data_df["Item Name"].unique())
AvePurPrice = purchase_data_df["Price"].mean()
TotalPurch = len(purchase_data_df["Purchase ID"])
TotalRev = purchase_data_df["Price"].sum()
```

```python
Purchase_Analysis_df = pd.DataFrame({
    "Unique Items": [NumUnique],
    "Average Purchase Price": [AvePurPrice],
    "Total Purchase": [TotalPurch],
    "Total Revenue": [TotalRev]
})


Purchase_Analysis_df["Average Purchase Price"] = Purchase_Analysis_df["Average Purchase Price"].map("${:.2f}".format)
Purchase_Analysis_df["Total Revenue"] = Purchase_Analysis_df["Total Revenue"].map("${:,}".format)

Purchase_Analysis_df
```

```python
# * Percentage and Count of Male Players
# * Percentage and Count of Female Players
# * Percentage and Count of Other / Non-Disclose
deduped_df = purchase_data_df.drop_duplicates(subset='SN', keep="first")
```

```python
Male = deduped_df["Gender"].value_counts()["Male"]
Female = deduped_df["Gender"].value_counts()["Female"]
Other = deduped_df["Gender"].value_counts()["Other / Non-Disclosed"]
TotalGen = deduped_df["Gender"].count()
Male_Ave = (Male/TotalGen)*100
Female_Ave =(Female/TotalGen)*100
Other_Ave = (Other/TotalGen)*100
```

```python
#ask why this line works like this with "" Figured out this left column header blank
Gender_Data_df = pd.DataFrame({"": ['Male', 'Female', 'Other / Non-Disclosed'],
                            "Percentage of Players": [Male_Ave, Female_Ave, Other_Ave],
                            "Total Count": [Male, Female, Other]})
Gender_Data_df["Percentage of Players"] = Gender_Data_df["Percentage of Players"].map("{:.2f}%".format)
Gender_Data_df = Gender_Data_df.set_index('')
Gender_Data_df
```

```python
# * The below each broken by gender
#   * Purchase Count
#   * Average Purchase Price
#   * Total Purchase Value
#   * Average Purchase Total per Person by Gender
Gender_Purchase_df = purchase_data_df.groupby(["Gender"])
```

```python
#   * Purchase Count
#   * Average Purchase Price
#   * Total Purchase Value

PurchCount = Gender_Purchase_df["SN"].count()
AvePurch = Gender_Purchase_df['Price'].mean()
TotalPurch = Gender_Purchase_df["Price"].sum()

#   * Average Purchase Total per Person by Gender (MTM need to remove duplicates again.)
# Duplicate2 = purchase_data_df.drop_duplicates(subset='SN', keep="first")
# Grouped_deduped_df = Duplicate2.groupby(["Gender"])

AvePurPer = TotalPurch / Gender_Data_df["Total Count"]
```

```python
Gender_Totals_df = pd.DataFrame({"Number of Purchases": PurchCount,
                            "Average Purchase": AvePurch,
                                "Total Purchase Value" : TotalPurch,
                                "Average Purchase Per Person" : AvePurPer})
Gender_Totals_df["Average Purchase"] = Gender_Totals_df["Average Purchase"].map("${:.2f}".format)
Gender_Totals_df["Total Purchase Value"] = Gender_Totals_df["Total Purchase Value"].map("${:.2f}".format)
Gender_Totals_df["Average Purchase Per Person"] = Gender_Totals_df["Average Purchase Per Person"].map("${:.2f}".format)
Gender_Totals_df

#Check with Esme why numbers for per person don't match sample
```

```python
# Age Demographics
# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table
```

```python
AgeBins = [0, 9, 15, 19, 24, 29, 34, 39, 150]
AgeLabels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(purchase_data_df["Age"], bins = AgeBins, labels = AgeLabels)
```

```python
purchase_data_df["Age Bracket"] = pd.cut(purchase_data_df["Age"], bins = AgeBins, labels = AgeLabels)
purchase_data_df.head()
```

```python
AgeBracket_df = purchase_data_df.groupby(["Age Bracket"])
```

```python
AgeCounts = AgeBracket_df["SN"].count()
Totalplayers = purchase_data_df["SN"].count()
AgePercent = (AgeCounts / Totalplayers) * 100
```

```python
AgeAnal_df = pd.DataFrame({"Total Count": AgeCounts,
                            "Percentage of Players": AgePercent,
                            })
AgeAnal_df["Percentage of Players"] = AgeAnal_df["Percentage of Players"].map("{:.2f}%".format)
AgeAnal_df
```

```python
PurchCount2 = AgeBracket_df["SN"].count()
AvePurch2 = AgeBracket_df['Price'].mean()
TotalPurchValue2 = AgeBracket_df["Price"].sum()

Duplicate3 = purchase_data_df.drop_duplicates(subset='SN', keep="first")
Age_deduped_df = Duplicate3.groupby(["Age Bracket"])

# Duplicate2 = AgeBracket_df.drop_duplicates(subset='SN', keep="first")
# Age_deduped_df = Duplicate2.groupby(["Age Bracket"])

AvePurPer2 = Age_deduped_df["Price"].sum() / Age_deduped_df["SN"].count()
```

```python
Age_Totals_df = pd.DataFrame({"Number of Purchases": PurchCount2,
                            "Average Purchase Price": AvePurch2,
                                "Total Purchase Value" : TotalPurchValue2,
                                "Average Purchase Per Person" : AvePurPer2})
Age_Totals_df["Average Purchase Price"] = Age_Totals_df["Average Purchase Price"].map("${:.2f}".format)
Age_Totals_df["Total Purchase Value"] = Age_Totals_df["Total Purchase Value"].map("${:.2f}".format)
Age_Totals_df["Average Purchase Per Person"] = Age_Totals_df["Average Purchase Per Person"].map("${:.2f}".format)
Age_Totals_df
```
